```python
import requests

# API URL and headers
url = "https://api.sportradar.com/tennis/trial/v3/en/double_competitors_rankings.json?api_key=0aRv55dviakXsgmkMGeQQvn67qnWxeUj3dHGPFwx"
headers = {"accept": "application/json"}

# Make the GET request
response = requests.get(url, headers=headers)

# Parse the JSON response
data = response.json()

# Extract and display the timestamp
generated_at = data.get("generated_at", "N/A")
print(f"Generated At: {generated_at}\n")

# Loop through the rankings list
for ranking in data.get("rankings", []):
    print(f"Ranking Type ID: {ranking.get('type_id', 'N/A')}")
    print(f"Ranking Name: {ranking.get('name', 'N/A')}")
    print(f"Year: {ranking.get('year', 'N/A')}")
    print(f"Week: {ranking.get('week', 'N/A')}")
    print(f"Gender: {ranking.get('gender', 'N/A')}")
    print("=" * 50)
    
    # Loop through the competitor rankings
    for competitor_ranking in ranking.get("competitor_rankings", []):
        # Extract competitor ranking details
        Rank_id = competitor_ranking.get("id", "N/A")
        Rank = competitor_ranking.get("rank", "N/A")
        Movement = competitor_ranking.get("movement", "N/A")
        Points = competitor_ranking.get("points", "N/A")
        Competitions_played = competitor_ranking.get("competitions_played", "N/A")
        
        # Extract competitor details
        competitor = competitor_ranking.get("competitor", {})
        Competitor_id = competitor.get("id", "N/A")
        Name = competitor.get("name", "N/A")
        Country = competitor.get("country", "N/A")
        Country_code = competitor.get("country_code", "N/A")
        Abbreviation = competitor.get("abbreviation", "N/A")
        
        # Print details with updated column names
        print(f"  Rank_id: {Rank_id}")
        print(f"  Rank: {Rank}")
        print(f"  Movement: {Movement}")
        print(f"  Points: {Points}")
        print(f"  Competitions_played: {Competitions_played}")
        print(f"  Competitor_id: {Competitor_id}")
        print(f"  Name: {Name}")
        print(f"  Country: {Country}")
        print(f"  Country_code: {Country_code}")
        print(f"  Abbreviation: {Abbreviation}")
        print("-" * 50)
    print("=" * 50)
```
