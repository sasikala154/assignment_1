```python
import requests

# API endpoint
url = "https://api.sportradar.com/tennis/trial/v3/en/complexes.json?api_key=0aRv55dviakXsgmkMGeQQvn67qnWxeUj3dHGPFwx"
headers = {"accept": "application/json"}

# Fetch data from the API
response = requests.get(url, headers=headers)

# Parse JSON response
data = response.json()

# Extract the "generated_at" field
generated_at = data["generated_at"]
print(f"Generated At: {generated_at}\n")

# Loop through the "complexes" in the JSON response
for complex_item in data["complexes"]:
    # Extract complex details
    Complex_id = complex_item["id"]
    Complex_name = complex_item["name"]
    print(f"Complex_id: {Complex_id}")
    print(f"Complex_name: {Complex_name}")
    
    # Extract venue details for each complex
    for venue in complex_item.get("venues", []):
        Venue_id = venue["id"]
        Venue_name = venue["name"]
        City_name = venue.get("city_name", "N/A")
        Country_name = venue.get("country_name", "N/A")
        Country_code = venue.get("country_code", "N/A")
        Timezone = venue.get("timezone", "N/A")
        
        # Print the venue details
        print(f"  Venue_id: {Venue_id}")
        print(f"  Venue_name: {Venue_name}")
        print(f"  City_name: {City_name}")
        print(f"  Country_name: {Country_name}")
        print(f"  Country_code: {Country_code}")
        print(f"  Timezone: {Timezone}")
        print("-" * 50)
    print("=" * 50)
```
