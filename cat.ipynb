```python
import requests

# API URL and headers
url = "https://api.sportradar.com/tennis/trial/v3/en/competitions.json?api_key=0aRv55dviakXsgmkMGeQQvn67qnWxeUj3dHGPFwx"
headers = {"accept": "application/json"}

# Fetch data from the API
response = requests.get(url, headers=headers)
data = response.json()

# Extract and print the "generated_at" field
generated_at = data["generated_at"]
print(f"Generated At: {generated_at}\n")

# Loop through the competitions and extract the required information
for competition in data["competitions"]:
    # Extract fields with new column names
    Category_id = competition["category"]["id"]
    Category_name = competition["category"]["name"]
    Competition_id = competition["id"]
    Competition_name = competition["name"]
    Parent_id = competition.get("parent_id", "N/A")
    Name = competition["name"]
    Gender = competition.get("gender", "N/A")
    Type = competition["type"]  # Newly added column "Type"

    # Print the extracted values
    print(f"Category_id: {Category_id}")
    print(f"Category_name: {Category_name}")
    print(f"Competition_id: {Competition_id}")
    print(f"Competition_name: {Competition_name}")
    print(f"Parent_id: {Parent_id}")
    print(f"Name: {Name}")
    print(f"Gender: {Gender}")
    print(f"Type: {Type}")
    print("-" * 50)   
```
